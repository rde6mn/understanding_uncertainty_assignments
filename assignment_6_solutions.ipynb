0, & x \le 0 \\
\sqrt{x}, & 0 \le x \le 1 \\

\end{cases}

As $x$ goes less than 0, we get $F(x)=0$, and for $x$ greater than 1 we get $F(x)=1$, so the boundary conditions that $\lim_{x \rightarrow -\infty}F(x)=0$ and $\lim_{x\rightarrow +\infty} F(x)=1$. We just have to test if it's increasing. Let's take a derivative for $x$ between 0 and 1:
$$
F'(x) = \frac{1}{2\sqrt{x}}>0
$$
So we have a continuous function increasing from 0 to 1: It can be a cdf. What's the pdf? For $x$ between 0 and 1, $
F'(x) = \frac{1}{2\sqrt{x}} = f(x),
$
and otherwise 0.

2. $F(x) = \dfrac{1}{1+e^{-x}}$

Let's remember, $ \lim_{x \rightarrow +\infty } e^{-x} = 0$ and $ \lim_{x \rightarrow -\infty } e^{-x} = \infty$. That means that $ \lim_{x \rightarrow +\infty } 1/(1+e^{-x}) = 1$ and 
$ \lim_{x \rightarrow -\infty } 1/(1+e^{-x}) = 0$. So we have the right boundary conditions for $F(x)$ to be a CDF. Let's compute the pdf and check that it's positive, so that $F(x)$ is non-decreasing
$$
F'(x) = \dfrac{e^{-x}}{(1+e^{-x})^2} = f(x) >0.
$$
So this could be a CDF.

3. For $ a < c < b$,
$
F(x) = \begin{cases}
0, & x \le 0 \\
\frac{(x-a)^2}{(b-a)(c-a)}, & a \le x \le b \\
1 - \frac{(b-x)^2}{(b-a)(b-c)}, & b < x < c \\
1, & x \ge b
\end{cases}
$

Taking a derivative yields
$$
f(x) = 
\begin{cases}
0, & x \le 0 \\
\frac{2(x-a)}{(b-a)(c-a)}, & a \le x \le b \\
\frac{2(b-x)}{(b-a)(b-c)}, & b < x < c \\
0, & x \ge b
\end{cases}
$$
which is non-negative in each case, so this is a non-decreasing function. The CDF goes to zero as $x$ goes to $-\infty$ and to 1$ as $x\rightarrow +\infty$. So this is a valid distribution function, called the "triangle distribution".

```python
def triangle_cdf(x,a=0,b=2,c=1):
    y = ( (x <= a)*0
          +(x>a)*(x<=c)* ((x-a)**2)/( (b-a)*(c-a))
          +(x>c)*(x<=b)* (1-((b-x)**2)/( (b-a)*(b-c)))
          +(x>b)*1
        )
    return y

def triangle_pdf(x,a=0,b=2,c=1):
    y = ( (x <= a)*0
          +(x>a)*(x<=c)* ( 2*(x-a))/( (b-a)*(c-a))
          +(x>c)*(x<=b)* (2*(b-x))/( (b-a)*(b-c))
          +(x>b)*0
        )
    return y

grid = np.linspace(-1,3,50)
cdf = triangle_cdf(grid)
pdf = triangle_pdf(grid)

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.lineplot(x=grid,y=pdf,ax=axes[0])
axes[0].set_title('Density')
sns.lineplot(x=grid,y=cdf,ax=axes[1])
axes[1].set_title('Distribution')
plt.show()
```

4. $ F(x) = \begin{cases}
0, & x <0 \\
1 - e^{-x}, & x>0
\end{cases}
$

Since $1-e^{-x}$ goes to $1-e^{0} =1-1=0$ as $x$ gets close to zero, this function goes to zero as $x$ gets small. Likewise, as $x \rightarrow \infty$, $1-e^{-x} \rightarrow 1-e^{-\infty} = 1-0$, so converges to 1. In between, the density is
$$
f(x) = \begin{cases}
0, & x <0 \\
e^{-x}, & x>0
\end{cases}
$$
which is non-negative, so the CDF is non-decreasing. This is a valid distribution function, called the Exponential Distribution.

```python
grid = np.linspace(0,5,20)
cdf = 1-np.exp(-grid)
pdf = np.exp(-grid)

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.lineplot(x=grid,y=pdf,ax=axes[0])
axes[0].set_title('Density')
sns.lineplot(x=grid,y=cdf,ax=axes[1])
axes[1].set_title('Distribution')
plt.show()
```


- Logistic distribution (similar to normal): 
$$
F(x; \sigma) = \dfrac{1}{1+e^{-x/\sigma}} 
$$
with $\sigma >0$.

The density is
$$
F'(x) = f(x) = \frac{1}{\sigma} \dfrac{e^{-x/\sigma}}{( 1+e^{-x/\sigma})^2} =\frac{1}{\sigma} \dfrac{e^{-x/\sigma}}{1+e^{-x/\sigma}} \frac{1}{ 1+e^{-x/\sigma}} =\frac{1}{\sigma}(1-F(x))(F(x))
$$
This is strictly positive for any real number, so the support of a logistically distributed random variable is all of $\mathbb{R} = (-\infty, +\infty)$.


```python
grid = np.linspace(-4,4,100)

## Plot the density and distribution
pdf = sp.stats.logistic.pdf(grid)
cdf = sp.stats.logistic.cdf(grid)

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.lineplot(x=grid,y=pdf,ax=axes[0])
axes[0].set_title('Density')
sns.lineplot(x=grid,y=cdf,ax=axes[1])
axes[1].set_title('Distribution')
plt.show()

# Sampling
sample = sp.stats.logistic.rvs(size=50000,random_state=10)

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.kdeplot(x=sample,ax=axes[0],label='kde')
sns.lineplot(x=grid,y=pdf,ax=axes[0],label='pdf',linestyle='--')
axes[0].set_title('Sample KDE')
sns.ecdfplot(x=sample,ax=axes[1],label='ecdf')
sns.lineplot(x=grid,y=cdf,ax=axes[1],label='cdf',linestyle='--')
axes[1].set_title('Sample ECDF')
plt.show()
```

- Exponential distribution (similar to log-normal):
$$
F_X(x) = \begin{cases}
0, & x<0 \\
1 - e^{-\lambda x}, & x \ge 0,
\end{cases}
$$
with $\lambda > 0$.


As $x \rightarrow \infty$, $e^{-\lambda x} \rightarrow e^{-\infty} = 0$, so $ 1-e^{-\lambda x} \rightarrow 1$. Likewise, as $x \rightarrow -\infty$, the CDF goes to 0 (it goes to zero at $x=0$). Taking a derivative,
$$
F_X'(x) = \lambda e^{-\lambda x} = f_X(x) >0,
$$
which gives the density and shows that it is non-decreasing. The support is all positive real numbers.

An example from METABRIC would be survival duration:



- Negative Binomial (similar to Poisson): The probability mass function for positive integers is:
$$
f(k;r,p) = \dfrac{(k+r-1)!}{k!(r-1)!}(1-p)^k p^r, \quad \text{ for $k=0,...n$}
$$
You can interpret this as follows: Flip a coin that comes up heads with probability $p$ until you get $r$ heads, and then stop. What is the probability of stopping at each $k=0,1,2,...$?


```python
grid = np.arange(20)

## Plot the density and distribution
pmf = sp.stats.nbinom.pmf(k=grid,n=5,p=.6)
cdf = sp.stats.nbinom.cdf(k=grid,n=5,p=.6)

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.barplot(x=grid,y=pmf,ax=axes[0])
axes[0].set_title('Density')
sns.lineplot(x=grid,y=cdf,ax=axes[1])
axes[1].set_title('Distribution')
plt.show()

# Sampling
sample = sp.stats.nbinom.rvs(n=5,p=.6,size=50000,random_state=10)

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.histplot(x=sample,ax=axes[0],stat='proportion',label='histogram')
axes[0].set_title('Sample KDE')
sns.ecdfplot(x=sample,ax=axes[1],label='ecdf')
sns.lineplot(x=grid,y=cdf,ax=axes[1],label='cdf',drawstyle='steps-post')
axes[1].set_title('Sample ECDF')
plt.show()
```




- Categorical (similar to Bernoulli): The probability mass function over $k = 1, 2, ..., K$ categories is
$$
f(k;p_1,...,p_K) = p_1^{k=1}p_2^{k=2}...p_K^{k=K}
$$
where $0 \le p_i \le 1$ and $ \sum_{k=1}^K p_k = 1$.


We want to give this one a name and recognize it's an option, but because the labels aren't necessarily ordered or even numeric, this is a bit of an odd option. However, for lots of data that is neither ordered not numeric, it's a useful thing to know. In order to give you a useful example, I'm going to combine chemotherapy, hormone therapy, and radiation therapy into one overall variable, and plot it:

```python
chemo = df['Chemotherapy'].replace({'NO':'No Chemo', 'YES':'Chemo'})
radio = df['Radio Therapy'].replace({'NO':'No Radio', 'YES':'Radio'})
hormone = df['Hormone Therapy'].replace({'NO':'No Hormone', 'YES':'Hormone'})
treat = chemo+', '+radio+', '+hormone
sns.histplot(treat, stat='proportion')
plt.xticks(rotation=90)
```

This is the PMF. This is the probability mass function for treatments. So, you can see a few things: Most people get radio and hormone. Few people get chemo and hormone or just chemo. The proportion of people who get none of these therapies is only about .15. This quickly gives us a sense of the overall treatment status of the population, and highlights the most and least likely combinations of treatments.
